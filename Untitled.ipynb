```python
from urllib import request
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
url = "http://www.gutenberg.org/files/74/74-0.txt"
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import operator
```

```python
response = request.urlopen(url)
raw = response.read().decode('utf8')
type(raw)
```

```python
len(raw)
```

```python
tokens=nltk.word_tokenize(raw)
len(tokens)
```

```python
types=set(tokens)
print(len(types))
```

```python
TTR=len(types)/len(tokens)
print(TTR)
```

# Tokens=91299
# Types=9278
# TTR = 0.1016

```python
d={}
for x in tokens:
    d.setdefault(x,0)
    d[x]+=1
sorted_d = sorted(d.items(), key=operator.itemgetter(1),reverse=True)
for x in sorted_d:
    print(x)
    print("\n")
    
    
```

# Test for Zipf's law:

```python
y=[]
x=[]
for z in sorted_d:
    x.append(len(z[0]))
    y.append(z[1])
```

```python
fig = plt.figure()
ax = plt.axes()


ax.plot(x,y)
plt.title("Word Length vs Frequency plot")
plt.xlabel("Word Length")
plt.ylabel("Frequency");
```

## Zipf's Law for length seems to hold true from above plot. We can clearly see as the length of words increases, Frequency of their appearance in Corpus gets decreased.

## Zipf's law for meanings seems to hold true most of the time for content words but not that much applicable over function words
