# Q.1 and Q.2 Downloading dataset and removing unnecessary symbols.

```python
response = request.urlopen(url)
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktParameters, PunktLanguageVars
raw = response.read().decode('utf8')
text=raw.replace('\n',' ').replace('\r',' ').replace('\t',' ').replace(',',' ').replace("''"," ")
```

#  Unigram, bigram, trigrams and quadgrams.

```python
from nltk.util import ngrams
from collections import Counter
token = nltk.word_tokenize(text)
token = nltk.word_tokenize(text)

unigrams = ngrams(text.split(), 1)

bigrams = ngrams(text.split(), 2)
trigrams = ngrams(text.split(), 3)
quadgrams = ngrams(text.split(), 4)

uf=dict()
vocab=0
for x in unigrams:
    if x not in uf:
        uf[x]=1
        vocab+=1
    else:
        uf[x]+=1
        vocab+=1

bf=dict()
bfl=0
for x in bigrams:
    if x not in bf:
        bf[x]=1
        bfl+=1
    else:
        bf[x]+=1
        bfl+=1

tf=dict()
for x in trigrams:
    if x not in tf:
        tf[x]=1
    else:
        tf[x]+=1

qf=dict()
for x in quadgrams:
    if x not in qf:
        qf[x]=1
    else:
        qf[x]+=1


```

# MLE for Unigram(probabilities of individual words)


```python
freq_1gram = nltk.FreqDist(token)

len_corpus = len(token)
def unigram_prob(word):

    return freq_1gram[ word] / len_corpus
for x in token:
    print(x,unigram_prob(x))

```

# MLE for bigram 

```python
bmle={}
for x in bf:
    bmle[x]=bf[x]/uf[(x[0],)]
    
for x in bmle:
    print(x,bmle[x])
```

# MLE for trigram 


```python
tmle={}
for x in tf:
    tmle[x]=tf[x]/bf[(x[0],x[1])]
tmle
```

# MLE for quadgram

```python
qmle={}
for x in qf:
    qmle[x]=qf[x]/tf[(x[0],x[1],x[2])]
qmle
```

# Q.5 Add-1 Smoothing

```python
addone={}
for x in bf:
    addone[x]=(bf[x]+1)/(uf[(x[0],)]+vocab)
for x in bf:
    print(x)
    print("Bigram MLE before smoothing : %f"%bf[x])
    print("After addone smoothing : %f"%addone[x])
```

### Significant change can be seen after Add-1 Smoothing.

# Q.6 Good-turing smoothing 

```python
fd={}
for x in bf.values():
    if x not in fd:
        fd[x]=1
    else:
        fd[x]+=1
fd[0]=bfl

gts={}
for i in range(10,-1,-1):
    gts[i]=((i + 1) * fd[i + 1]) / fd[i]

for i in range(10):
    print(i,gts[i],i-gts[i])
```
